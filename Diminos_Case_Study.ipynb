###
The analysis shows that while the average delivery time is consistently low (around 17–18 minutes), Diminos evaluates store performance using the 95th percentile delivery time, not the average. This means that 95% of all orders must be delivered within 31 minutes to comply.

Current results indicate that the store is meeting the cutoff overall, but a small percentage of late deliveries during specific peak hours have the potential to push the 95th percentile above 31 minutes, which would put the franchise at risk.These delays are not visible in average-based metrics and only become apparent when percentile-based analysis is performed

Day-wise breakdown reveals that Tuesday consistently shows the highest 95th percentile delivery time, placing it in a risk zone. This indicates that a higher proportion of late deliveries occur on Tuesdays compared to other days
